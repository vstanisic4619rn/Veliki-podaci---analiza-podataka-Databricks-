```
## Trazimo osvajaca lige tako sto racunamo poene za sve
 
df15_1 = spark.sql("select home as home1, count(home) as wins_1, count(home) * 3 as points_1 from df15_teams where result = '1' group by home order by count(home) desc")
df15_x1 = spark.sql("select home as homex1, count(home) as egal_x1, count(home) as points_x1 from df15_teams where result = 'x' group by home order by count(home) desc")
df15_2 = spark.sql("select away as away2, count(away) as wins_2, count(away) * 3 as points_2 from df15_teams where result = '2' group by away order by count(away) desc")
df15_x2 = spark.sql("select away as awayx2, count(away) as egal_x2, count(away) as points_x2 from df15_teams where result = 'x' group by away order by count(away) desc")
 
df15_sum = df15_1.join(df15_x1, [df15_1.home1 == df15_x1.homex1]).join(df15_2, [df15_1.home1 == df15_2.away2]).join(df15_x2, [df15_1.home1 == df15_x2.awayx2])
df15_sum = df15_sum.withColumn("points", df15_sum.points_1 + df15_sum.points_x1 + df15_sum.points_2 + df15_sum.points_x2).sort(col("points").desc())
 
print("2015")
df15_sum = df15_sum.select("home1", "points")
df15_sum.show()
 
df20_1 = spark.sql("select home as home1, count(home) as wins_1, count(home) * 3 as points_1 from df20_teams where result = '1' group by home order by count(home) desc")
df20_x1 = spark.sql("select home as homex1, count(home) as egal_x1, count(home) as points_x1 from df20_teams where result = 'x' group by home order by count(home) desc")
df20_2 = spark.sql("select away as away2, count(away) as wins_2, count(away) * 3 as points_2 from df20_teams where result = '2' group by away order by count(away) desc")
df20_x2 = spark.sql("select away as awayx2, count(away) as egal_x2, count(away) as points_x2 from df20_teams where result = 'x' group by away order by count(away) desc")
 
df20_sum = df20_1.join(df20_x1, [df20_1.home1 == df20_x1.homex1]).join(df20_2, [df20_1.home1 == df20_2.away2]).join(df20_x2, [df20_1.home1 == df20_x2.awayx2])
df20_sum = df20_sum.withColumn("points", df20_sum.points_1 + df20_sum.points_x1 + df20_sum.points_2 + df20_sum.points_x2).sort(col("points").desc())
 
print("2020.")
df20_sum = df20_sum.select("home1", "points")
df20_sum.show()
 
df15_sum.createOrReplaceTempView("rank15")
df20_sum.createOrReplaceTempView("rank20")

```

```
from pyspark.sql.types import *

inputPath = "dbfs:/df15_csv"

# Since we know the data format already, let's define the schema to speed up processing (no need for Spark to infer schema)
jsonSchema = StructType([ StructField("time", TimestampType(), True), StructField("action", StringType(), True) ])

# Static DataFrame representing data in the JSON files
staticInputDF = (
  spark
    .read
    .schema(jsonSchema)
    .json(inputPath)
)

display(staticInputDF)
```

```
from pyspark.sql.functions import *      # for window() function

staticCountsDF = (staticInputDF.groupBy(staticInputDF.action, window(staticInputDF.time, "1 hour")).count())
staticCountsDF.cache()

display(staticCountsDF)
```

```
# Register the DataFrame as table 'static_counts'
staticCountsDF.createOrReplaceTempView("static_counts")
```

```
%sql select action, sum(count) as total_count from static_counts group by action
```

```
%sql select action, date_format(window.end, "MMM-dd HH:mm") as time, count from static_counts order by time, action
```

```
from pyspark.sql.functions import *

# Similar to definition of staticInputDF above, just using `readStream` instead of `read`
streamingInputDF = (
  spark
    .readStream                       
    .schema(jsonSchema)               # Set the schema of the JSON data
    .option("maxFilesPerTrigger", 1)  # Treat a sequence of files as a stream by picking one file at a time
    .json(inputPath)
)

# Same query as staticInputDF
streamingCountsDF = (streamingInputDF.groupBy(streamingInputDF.action, window(streamingInputDF.time, "1 hour")).count())

# Is this DF actually a streaming DF?
streamingCountsDF.isStreaming
```

```
spark.conf.set("spark.sql.shuffle.partitions", "2")  # keep the size of shuffles small

query = (
  streamingCountsDF
    .writeStream
    .format("memory")        # memory = store in-memory table 
    .queryName("counts")     # counts = name of the in-memory table
    .outputMode("complete")  # complete = all the counts should be in the table
    .start()
)
```
